```python
import pandas as pd
from plotly.offline import init_notebook_mode, iplot, plot
# import plotly as py
init_notebook_mode(connected=True)
import plotly.graph_objs as go
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("F:\Machine learning\Regression\E-commerce\clothes dataset (4)\Ecommerce Customers.csv")
df.sample(5)
```

```python
df.drop(['Email' , 'Avatar' , 'Address'] , axis = 1 , inplace = True)
df.head()
```

```python
for col in df.columns:
    fig = go.Figure(
        data=[
            go.Box(
                y=df[col],
                name = col,
                marker=dict(
                    color='rgb(12, 12, 140)'
                )
            )
        ]
    )
    iplot(fig)
```
