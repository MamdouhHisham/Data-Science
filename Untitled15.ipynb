# Linear Regression from Scratch

This notebook demonstrates the implementation of a linear regression model from scratch.

## Steps of the Algorithm

1. Set the learning rate and number of iterations, and initialize random weight and bias values.
2. Build the linear regression equation \( y = wx + b \).
3. Calculate the predicted \( y \) value (`Y_pred`) for a given \( X \) value using the current weight and bias.
4. Compute the loss function for these parameter values (difference between `Y_pred` and `Y_true`).
5. Update the parameter values using gradient descent (new weight and bias values).
6. Repeat steps 3, 4, and 5 until the loss function is minimized, yielding the best model (optimal weight and bias values).

## Libraries Used
1. Numpy
2. Pandas for manipulate dataset
3. Plotly for visualizing the dataset

```python
import numpy as np
import random
```

```python
class LinearRegression:
    def __init__(self, learning_rate, num_of_iterations):
        self.learning_rate = learning_rate
        self.num_of_iterations = num_of_iterations

    def fit(self, X, Y):
        n_samples, n_features = X.shape

        self.weights = np.random.randn(n_features)
        self.bias = 0

        iterations = []
        costs = []
        
        for i in range(self.num_of_iterations):
            y_predicted = np.dot(X, self.weights) + self.bias

            dw = (1 / n_samples) * np.dot(X.T, (y_predicted - Y))

            if np.isnan(dw).any() or np.isinf(dw).any():
              dw = np.where(np.isnan(dw), 0, dw)
              dw = np.where(np.isinf(dw), 0, dw)

            db = (1 / n_samples) * np.sum(y_predicted - Y)

            if np.isnan(db) or np.isinf(db):
              db = 0 

            iterations[i]=i
            costs[i] = np.mean((Y - y_predicted) ** 2)
           
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

        return iterations, costs

    def predict(self, X):
        return np.dot(X, self.weights) + self.bias
```
